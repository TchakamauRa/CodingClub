# Matplotlib Basics

Matplotlib is a high-level plotting package that provides a powerful and flexible interface for visualizing data.
Matplotlib was originally designed as an implementation of `matlab`'s plotting routines in `python` but has since become a unique widely used tool in its own right.

**Additional Recommended Resources:**
- [Matplotlib Documentation](https://matplotlib.org/tutorials/index.html)

In this tutorial, we'll cover the basics of how to visualize data through a variety of methods.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Similar to how we import `numpy` under the `np` namespace, it is common to import `matplotlib.pyplot` under the `plt` namespace.
Note that this does not import _all_ of Matplotlib, but does give us most of the functionality that we will regularly use.

## Scatter Plots

```python
x = np.linspace(0, 1, 10)
y = np.exp(-x)
```

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
```

## Line Plots

```python
x = np.linspace(0, 1, 10)
y= np.exp(-x)
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot')
```

## 1D and 2D Histograms

```python
np.random.seed(0)
n = 100000
x = np.random.standard_normal(n)
y = 2.0 + 3.0 * x + 4.0 * np.random.standard_normal(n)
```

```python
plt.hist(y)
plt.xlabel('x')
plt.ylabel('counts')
plt.title('1D Histogram')
```

```python
plt.hist(y)
plt.xlabel('y')
plt.ylabel('counts')
plt.title('1D Histogram')
```

```python
plt.hist2d(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Histogram')
```

## Contour Plots

```python
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-X**2 - Y**2)
```

```python
plt.contour(X, Y, Z)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Unfilled Contour Plot')
```

```python
plt.contourf(X, Y, Z)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Filled Contour Plot')
```

## Remarks

If you've been looking closely, you may have noticed that I redefined the variables `x` and `y` several times.
Indeed, if you go back and try to render the first few plots (without first executing the corresponding cells where `x` and `y` are defined), they will look different than before because our variables have been given new values.
This concept is called state, and the hidden state of these variables can be very confusing in `Jupyter` notebooks.
It's always careful to run through all of your cells in order so you don't end up with results that confuse you; remember, reproducability is important!

Try running _just_ the plot cells from the earlier examples and you'll get a better sense for how this works.
