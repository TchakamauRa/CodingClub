# Visualization

In this tutorial, we'll cover how to visualize data through a variety of methods.

To do so, we will use the `matplotlib` package, which was originally designed as an implementation of `matlab`'s plotting procedures in `python` but has since become more flexible and powerful.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('example_data.csv', delimiter=',', names=True)
```

## Scatter Plots

```python
plt.scatter(data['t'], data['x'], label='x')
plt.scatter(data['t'], data['y'], label='y')
plt.scatter(data['t'], data['z'], label='y')
plt.legend(loc='upper left')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.title('Position of a Particle')
```

## Line Plots

```python
plt.plot(data['t'], data['x'])
```

```python
plt.plot(data['t'], data['x'])
plt.plot(data['t'], data['y'])
plt.plot(data['t'], data['z'])
```

```python
plt.plot(data['t'], data['x'], label='x')
plt.plot(data['t'], data['y'], label='y')
plt.plot(data['t'], data['z'], label='y')
plt.legend(loc='upper left')
```

```python
plt.plot(data['t'], data['x'], label='x')
plt.plot(data['t'], data['y'], label='y')
plt.plot(data['t'], data['z'], label='y')
plt.legend(loc='upper left')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.title('Position of a Particle')
```

## 1D and 2D Histograms

```python
plt.hist2d(data['x'], data['y'])
```

## Contour Plots

```python
plt.contourf(np.array([data['x'], data['y']]))
```
